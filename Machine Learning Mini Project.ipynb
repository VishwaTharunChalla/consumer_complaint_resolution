## Task-25
## Scaling the Data Sets (note: discard dependent variable before doing standardization) and Make feature Selection with the help of PCA up to 80% of the information.

```python
# Removing dependent variable from the dataset
x=final_dummy_data.drop(columns='Consumer disputed?')
y=final_dummy_data['Consumer disputed?']
```

## Covariance (or) Correlation

```python
import numpy as np
cor_mat=np.corrcoef(x,rowvar=False)
```

## Decomposition

```python
eigen_values,eigen_vectors=np.linalg.eig(cor_mat)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
x.shape
```

```python
eigen_values.sum()
```

```python
sorted_eig=sorted(eigen_values,reverse=True)
sorted_eig
```

```python
pc_data=pd.DataFrame({'PC':['PC{}'.format(i) for i in range(1,len(sorted_eig)+1)],
                     'var_cap_units':np.round(sorted_eig,2),
                      'var_cap_perc':np.round(sorted_eig/sum(sorted_eig)*100,2),
                      'cumulative_var_cap':np.cumsum(np.round(sorted_eig/sum(sorted_eig)*100,2))})
```

```python
pc_data=pc_data[:30]
pc_data
```

```python
## Scree Plot:- It is a line plot of the eigen values (or) principal components in an analysis.
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.bar(x=pc_data.PC,height=pc_data.var_cap_perc)
plt.step(x=pc_data.PC,y=pc_data.cumulative_var_cap)
plt.show()
```

## select the no of PCs (sufficient for analysis)

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=11)   # above plot up to PC6 is is consistently plotted
```

```python
# Standardization the dataset using StandardScaler()
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x)
```

```python
x_std=sc.transform(x)
```

```python
final_pc=pca.fit_transform(x_std)
final_pc
```

```python
pca.explained_variance_ratio_
```

```python
final_pc_data=pd.DataFrame(pca.fit_transform(x_std),columns=['PC{}'.format(i) for i in range(1,12)])
final_pc_data['Consumer disputed?']=y
final_pc_data
```

## K-Nearest Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
acc,err=[],[]
for k in range(1,35):
    mod=KNeighborsClassifier(n_neighbors=k)
    mod.fit(x_train,y_train)
    y_pred=mod.predict(x_test)
    accuracy=accuracy_score(y_true=y_test,y_pred=y_pred)
    acc.append(accuracy)
    err.append(1-accuracy)
```

```python
f,axes=plt.subplots(1,2,figsize=(25,5))
axes[0].plot(range(1,35),acc)
axes[0].set_title('Accuracy')
axes[1].plot(range(1,35),err)
axes[1].set_title('Error Rate')
plt.show()
```
